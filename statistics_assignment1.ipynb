```python
import matplotlib.pyplot as plt
plt.figure(figsize=(3,3))
plt.scatter(mdm['Age'],mdm['degree of anisotropy'],c='Green',marker='*',s=16)
plt.xlabel('Age',c='Blue')
plt.ylabel('degree of anisotropy',c='Blue')
plt.title('Age Vs DOA for Males',c='Red')
plt.show()
print(mdm.corr())
print(mdm.corr(method="spearman"))
print(mdm.corr(method="kendall"))
```

```python
import pandas as pd
bp = pd.read_csv('A1Q1a.dat',names=['Volume of AV malformation in the brain'])
plt.figure(figsize=(4,4))
plt.boxplot(bp['Volume of AV malformation in the brain'])
plt.ylabel('Volume(mm)',c='Blue')
plt.title('Volume(mm) of AV malformation in the brain',c='Red')
plt.show()
print(bp['Volume of AV malformation in the brain'].mean())
print(bp['Volume of AV malformation in the brain'].median())
print(bp['Volume of AV malformation in the brain'].mode())
print(bp['Volume of AV malformation in the brain'].var())
print(bp['Volume of AV malformation in the brain'].std())
bp.describe()
```

```python
import pandas as pd
mdb = pd.read_csv('A1Q2b.dat',sep=' ',names =['Hemoglobin(in g/dl)','hematocrit levels'])
import matplotlib.pyplot as plt
plt.figure(figsize=(5,5))
plt.scatter(mdb['Hemoglobin(in g/dl)'],mdb['hematocrit levels'],c='Green',marker='*',s=16)
plt.xlabel('Hemoglobin(in g/dl)',c='Blue')
plt.ylabel('hematocrit levels',c='Blue')
plt.title('Hemoglobin (in g/dl) and hematocrit levels (in %) of dengue patients',c='Red')
plt.show()
print(mdb.corr())
print(mdb.corr(method="spearman"))
print(mdb.corr(method="kendall"))
```
